## 2) Metrics helpers

**FP/FN definition (binary anomaly):**
- anomaly = `severity != 'Normal'`
- **FP**: true Normal but predicted anomaly
- **FN**: true anomaly but predicted Normal

```python
import numpy as np
import time

def binary_fp_fn(y_true_sev, y_pred_sev):
    yt = np.array([0 if s == "Normal" else 1 for s in y_true_sev])
    yp = np.array([0 if s == "Normal" else 1 for s in y_pred_sev])
    tn = int(np.sum((yt == 0) & (yp == 0)))
    tp = int(np.sum((yt == 1) & (yp == 1)))
    fp = int(np.sum((yt == 0) & (yp == 1)))
    fn = int(np.sum((yt == 1) & (yp == 0)))
    fpr = fp / max(fp + tn, 1)
    fnr = fn / max(fn + tp, 1)
    return {"TP": tp, "TN": tn, "FP": fp, "FN": fn, "FPR": fpr, "FNR": fnr}

def mean_ms(func, n_runs=200):
    t0 = time.perf_counter()
    for _ in range(n_runs):
        func()
    return (time.perf_counter() - t0) * 1000 / n_runs
```

## 3) Approach A — Machine Learning (Random Forest)

train **two** models:
- Model 1: predict **severity**
- Model 2: predict **correction action**


```python
X = df[FEATURES].values

le_sev = LabelEncoder().fit(df[TARGET_SEV])
le_act = LabelEncoder().fit(df[TARGET_ACT])

y_sev = le_sev.transform(df[TARGET_SEV])
y_act = le_act.transform(df[TARGET_ACT])

X_train, X_test, ys_tr, ys_te, ya_tr, ya_te = train_test_split(
    X, y_sev, y_act, test_size=0.25, random_state=42, stratify=y_sev
)

clf_sev = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42, class_weight='balanced')
clf_act = RandomForestClassifier(n_estimators=200, max_depth=12, random_state=42, class_weight='balanced')

clf_sev.fit(X_train, ys_tr)
clf_act.fit(X_train, ya_tr)

pred_sev_test = le_sev.inverse_transform(clf_sev.predict(X_test))
pred_act_test = le_act.inverse_transform(clf_act.predict(X_test))

true_sev_test = le_sev.inverse_transform(ys_te)
true_act_test = le_act.inverse_transform(ya_te)

print('Severity hold-out accuracy:', accuracy_score(true_sev_test, pred_sev_test))
print('Action hold-out accuracy:', accuracy_score(true_act_test, pred_act_test))
```

```python
# Full-dataset predictions (for side-by-side comparison with fuzzy/rules)
pred_sev_ml = le_sev.inverse_transform(clf_sev.predict(X))
pred_act_ml = le_act.inverse_transform(clf_act.predict(X))

print("ML severity accuracy (full data):", accuracy_score(df[TARGET_SEV], pred_sev_ml))
print("ML action accuracy (full data):", accuracy_score(df[TARGET_ACT], pred_act_ml))

print("\nSeverity report (ML):\n")
print(classification_report(df[TARGET_SEV], pred_sev_ml, zero_division=0))

print("Confusion matrix (severity) [Normal, Moderate, Severe]:")
print(confusion_matrix(df[TARGET_SEV], pred_sev_ml, labels=["Normal","Moderate","Severe"]))

print("Binary FP/FN (ML):", binary_fp_fn(df[TARGET_SEV], pred_sev_ml))
```

```python
# Inference time (ML): severity + action prediction
def ml_infer_all():
    _ = clf_sev.predict(X)
    _ = clf_act.predict(X)

ms_ml = mean_ms(ml_infer_all, n_runs=200)
print("ML mean inference time per full dataset (ms):", ms_ml)
print("ML approx per row (microseconds):", (ms_ml / len(df)) * 1000)
```

## 4) Approach B — Fuzzy Logic


- **Min–max normalisation** per sensor into [0,1]
- **Triangular membership** functions: Low / Medium / High
- An aggregated **anomaly degree** (0..1)
- Two thresholds (t1, t2) to map anomaly degree → severity
- Priority rules to recommend an action


```python
def tri(x, a, b, c):
    if x <= a or x >= c:
        return 0.0
    if x <= b:
        return (x - a) / (b - a + 1e-9)
    return (c - x) / (c - b + 1e-9)

mins = df[FEATURES].min()
maxs = df[FEATURES].max()

def norm(v, lo, hi):
    return float(np.clip((v - lo) / (hi - lo + 1e-9), 0.0, 1.0))

def fuzzify_norm(x):
    return {
        'L': tri(x, -0.05, 0.0, 0.38),
        'M': tri(x, 0.18, 0.48, 0.82),
        'H': tri(x, 0.52, 0.82, 1.05),
    }

def anomaly_degree(row):
    nd = norm(row['delay_ms'], mins['delay_ms'], maxs['delay_ms'])
    nt = norm(row['clutch_temp_c'], mins['clutch_temp_c'], maxs['clutch_temp_c'])
    nq = norm(row['torque_var_pct'], mins['torque_var_pct'], maxs['torque_var_pct'])
    nr = norm(row['rpm_diff'], mins['rpm_diff'], maxs['rpm_diff'])

    md = fuzzify_norm(nd)
    mt = fuzzify_norm(nt)
    mq = fuzzify_norm(nq)
    mr = fuzzify_norm(nr)

    high_term = 0.28*md['H'] + 0.27*mt['H'] + 0.23*mq['H'] + 0.22*mr['H']
    mid_term  = 0.28*md['M'] + 0.27*mt['M'] + 0.23*mq['M'] + 0.22*mr['M']
    low_term  = 0.28*md['L'] + 0.27*mt['L'] + 0.23*mq['L'] + 0.22*mr['L']

    return float(np.clip(high_term + 0.55*mid_term - 0.35*low_term, 0.0, 1.0))

degrees = df.apply(anomaly_degree, axis=1)
degrees.describe()
```

```python
# Calibrate severity thresholds (t1,t2) on the provided dataset
best = (-1.0, 0.2, 0.6)
for a in np.linspace(0.18, 0.45, 18):
    for b in np.linspace(a + 0.05, 0.85, 18):
        pred = []
        for d in degrees:
            if d < a:
                pred.append('Normal')
            elif d < b:
                pred.append('Moderate')
            else:
                pred.append('Severe')
        acc = accuracy_score(df[TARGET_SEV], pred)
        if acc > best[0]:
            best = (acc, float(a), float(b))

t1, t2 = best[1], best[2]
print('Fuzzy calibrated: acc=', best[0], 't1=', t1, 't2=', t2)

def fuzzy_severity(d):
    if d < t1:
        return 'Normal'
    if d < t2:
        return 'Moderate'
    return 'Severe'

def fuzzy_action(row, sev):
    if sev == 'Normal':
        return 'NO_ACTION'
    if row['clutch_temp_c'] >= 120:
        return 'CLUTCH_PRESSURE_RECALIB'
    if row['rpm_diff'] >= 165:
        return 'RPM_SYNC'
    if row['delay_ms'] >= 150:
        return 'SHIFT_TIMING_ADJUST'
    if row['torque_var_pct'] >= 9.0:
        return 'TORQUE_REDISTRIBUTION'
    return 'SHIFT_TIMING_ADJUST' if sev == 'Moderate' else 'TORQUE_REDISTRIBUTION'

pred_sev_fuzzy = [fuzzy_severity(d) for d in degrees]
pred_act_fuzzy = [fuzzy_action(r, s) for (_, r), s in zip(df.iterrows(), pred_sev_fuzzy)]

print('Fuzzy severity accuracy:', accuracy_score(df[TARGET_SEV], pred_sev_fuzzy))
print('Fuzzy action accuracy:', accuracy_score(df[TARGET_ACT], pred_act_fuzzy))
print('Binary FP/FN (Fuzzy):', binary_fp_fn(df[TARGET_SEV], pred_sev_fuzzy))
```

```python
def fuzzy_infer_all():
    deg = df.apply(anomaly_degree, axis=1)
    sev = [fuzzy_severity(d) for d in deg]
    act = [fuzzy_action(r, s) for (_, r), s in zip(df.iterrows(), sev)]
    return sev, act

ms_fuzzy = mean_ms(lambda: fuzzy_infer_all(), n_runs=60)
print('Fuzzy mean inference time per full dataset (ms):', ms_fuzzy)
print('Fuzzy approx per row (microseconds):', (ms_fuzzy / len(df)) * 1000)
```

##  Approach C — Rule-Based

```python
def rule_severity(row):
    delay, temp, torque, rpm = row['delay_ms'], row['clutch_temp_c'], row['torque_var_pct'], row['rpm_diff']
    score = 0
    score += 3 if delay >= 150 else 2 if delay >= 115 else 1 if delay >= 100 else 0
    score += 3 if temp >= 120 else 2 if temp >= 105 else 1 if temp >= 100 else 0
    score += 3 if torque >= 10.0 else 2 if torque >= 6.5 else 1 if torque >= 5.0 else 0
    score += 3 if rpm >= 160 else 2 if rpm >= 120 else 1 if rpm >= 90 else 0
    if score >= 6:
        return 'Severe'
    if score >= 3:
        return 'Moderate'
    return 'Normal'

def rule_action(row, sev):
    if sev == 'Normal':
        return 'NO_ACTION'
    delay, temp, torque, rpm = row['delay_ms'], row['clutch_temp_c'], row['torque_var_pct'], row['rpm_diff']
    if sev == 'Severe':
        if temp >= 120:
            return 'CLUTCH_PRESSURE_RECALIB'
        if rpm >= 165:
            return 'RPM_SYNC'
        if torque >= 11.0:
            return 'TORQUE_REDISTRIBUTION'
        if delay >= 150:
            return 'SHIFT_TIMING_ADJUST'
        return 'TORQUE_REDISTRIBUTION'
    # Moderate
    if delay >= 130 and torque < 8:
        return 'SHIFT_TIMING_ADJUST'
    if torque >= 6.0:
        return 'TORQUE_REDISTRIBUTION'
    if rpm >= 100:
        return 'TORQUE_REDISTRIBUTION'
    return 'SHIFT_TIMING_ADJUST'

pred_sev_rule = df.apply(rule_severity, axis=1).tolist()
pred_act_rule = [rule_action(r, s) for (_, r), s in zip(df.iterrows(), pred_sev_rule)]

print('Rule severity accuracy:', accuracy_score(df[TARGET_SEV], pred_sev_rule))
print('Rule action accuracy:', accuracy_score(df[TARGET_ACT], pred_act_rule))
print('Binary FP/FN (Rule):', binary_fp_fn(df[TARGET_SEV], pred_sev_rule))
```

```python
def rule_infer_all():
    sev = df.apply(rule_severity, axis=1).tolist()
    act = [rule_action(r, s) for (_, r), s in zip(df.iterrows(), sev)]
    return sev, act

ms_rule = mean_ms(lambda: rule_infer_all(), n_runs=200)
print('Rule mean inference time per full dataset (ms):', ms_rule)
print('Rule approx per row (microseconds):', (ms_rule / len(df)) * 1000)
```
